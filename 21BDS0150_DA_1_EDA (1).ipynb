```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/econmath.csv'
data = pd.read_csv(url)
print(data.head())
```

```python
# Get the dimension and summary
print(f"Dataset Dimensions: {data.shape}")
print(data.info())
print(data.describe())
```

```python
#Data Cleaning
data.drop(['rownames'], axis = 1, inplace = True)

# Check for missing values
print(data.isnull().sum())
```

```python
# Fill missing values with the median
data.fillna(data.median(), inplace=True)
```

```python
# Check for data types
print(data.dtypes)
```

```python
print(data.info())
print(data.head())
```

```python
# Univariate Analysis
# Loop through each column
for col in data.columns:
    print(f"--- Univariate Analysis for {col} ---")

    # Check if the column is numerical
    if data[col].dtype != 'object':
        print(f"Mean: {data[col].mean()}")
        print(f"Median: {data[col].median()}")
        print(f"Standard Deviation: {data[col].std()}")
        print(f"Variance: {data[col].var()}")
        print(f"Skewness: {data[col].skew()}")
        print(f"Kurtosis: {data[col].kurt()}")

    # Visualising Histogram and Box Plot for numerical features
    plt.figure(figsize=(10, 4))

    if data[col].dtype != 'object':
        plt.subplot(1, 2, 1)
        sns.histplot(data[col], kde=True, color='blue')
        plt.title(f"Histogram of {col}")

        plt.subplot(1, 2, 2)
        sns.boxplot(x=data[col], color='green')
        plt.title(f"Box Plot of {col}")

    plt.tight_layout()
    plt.show()
```

```python
# Bivariate Analysis
# Scatterplot for fathcoll vs mathcoll
plt.figure(figsize=(8, 6))
sns.scatterplot(x='colgpa', y='hsgpa', data=data, color='blue')
plt.title('Scatterplot of colgpa vs hsgpa')
plt.xlabel('colgpa')
plt.ylabel('hsgpa')
plt.show()
```

```python
# Multivariate Analysis
plt.figure(figsize= (8, 5))
sns.set(style="ticks", color_codes=True)
sns.pairplot(data, vars = ['colgpa', 'hsgpa','score'], kind = 'reg')
plt.show()
```

```python
# Correlation using Pearson Method
correlation_matrix = data.corr(method = 'pearson')
```

```python
print(correlation_matrix)
```

```python
# Multivariate Analysis
#Visualise using heatmap
plt.figure(figsize= (10,8))
sns.heatmap(correlation_matrix, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
